**TYPE** **CASTING**

We can convert one type value to another type. This conversion is called Typecasting or Type coersion.

֍ The following are various inbuilt functions for type casting.

int()

float()

complex()

֍ int():

We can use this function to convert values from other types to int

```python
print(int(34.32))
print(int(454+544j))
```

```python
print(int(False))
print("425")
```

 float(): We can use float() function to convert other type values to float type. (We can convert string value containing letters into float but if string value is number, we can convert it into a string)

```python
print(float(4535))
print(float(False))
print(float("123Test"))
```

```python
print(float(12+32j))
```

complex(): We can use complex() function to convert other types to complex type.

Form-1: complex(x)

We can use this function to convert x into complex number with real part x and imaginary part 0.

```python
print(complex(667.78))
print(complex(True))
print(complex(False))
print(complex(323))
print(complex("Test123"))
```

Form-2: complex(x,y)

We can use this method to convert x and y into complex number such that x will be real part and y will be imaginary part.

Eg: complex(10, -2)  10-2j

complex(True, False)  1+0j

֍ bool(): We can use this function to convert other type values to bool type.

If the value is zero the result is false otherwise result will be true

```python
print(bool(667.78))
print(bool(0))
print(bool(1))
print(bool(-12))
print(bool("Test123"))
print(bool(1+12j))
```

֍ str(): We can use this method to convert other type values to str type.



```python
print(str(667.78))
print(str(0))
print(str(1))
print(str(-12))
print(str("Test123"))
print(str(1+12j))
```

**Fundamental Data Types vs Immutability:**

֍ All Fundamental Data types are immutable. i.e once we creates an object,we cannot perform any changes in that object. If we are trying to change then with those changes a new object will be created. This non-chageable behaviour is called immutability.

```python
a = 123
b = 123
b = 323
print(a is b)
print(id(a))
print(id(b))
```

**6) bytes Data Type:**
bytes data type represens a group of byte numbers just like an array.

Conclusion 1: The only allowed values for byte data type are 0 to 256. By mistake if we are trying to provide any other values then we will get value error.

Conclusion 2: Once we creates bytes data type value, we cannot change its values,otherwise we will get TypeError.

```python
abc = [1,2,3,4,5]
print(type(abc))
ab = bytes(abc)
print(type(ab))

x = [1000,1200,2100]
y = bytes(x)
```

```python
xy = [10,11,12,13]
x = bytes(xy)
x[1] = 15
```

**7) bytearray Data Type:**
bytearray is exactly same as bytes data type except that its elements can be modified.

```python
a = [1,2,3,4,5]
b = bytearray(a)
b[3] = 11
print(b)
print(type(b))
for i in b:
  print(i)
```

**8) List Data Type:**

If we want to represent a group of values as a single entity where insertion order required to preserve and duplicates are allowed then we should go for list data type.

Insertion Order is preserved

Heterogeneous Objects are allowed

Duplicates are allowed

Growable in nature

Values should be enclosed within square brackets.

```python
list = [12, 12.2, True, "Data science", 12+1j]
print(list)
print(type(list))

# New item assignment is possible
list[1] = 100
print(list)

# It is growable in nature
list.append(1234)
print(list)

#removal of item is also possible
list.remove(100)
print(list)
list1 = list*2
print(list1)
```

**9) Tuple Data Type:**

 tuple data type is exactly same as list data type except that it is immutable.i.e we cannot chage values.

 Tuple elements can be represented within parenthesis.

 tuple is the read only version of list

```python
tuple = (1,2.1,3+2j,"Data",False)
print(type(tuple))
print(tuple)
tuple[2] = 12
print(tuple)
```

**Range Data Type:**

 range Data Type represents a sequence of numbers.

 The elements present in range Data type are not modifiable. i.e range Data type is immutable.

```python
print(range(4))

a = range(1,5) #generates numbers from 1 to 4
print(a)
for i in a:
  print(i)
```

```python
b = range(10,20,2)  #2 means increament value
for i in b:
  print(i)
```

```python
#We can access elements present in the range Data Type by using index
print(a[0])
print("-----------------")
#We can not modify the values of range data type by using below code.
a[0] = 100
```

```python
# We can create a list of values with range data type
list(range(6))
```

**11) set Data Type:**

֍ If we want to represent a group of values without duplicates where order is not important then we should go for set Data Type.

- Insertion order is not preserved
- Duplicates are not allowed
- Heterogeneous objects are allowed
- Index concept is not applicable
- It is mutable collection
- Growable in nature

```python
s = {100,0,10,200,10,'Data Science',12.12,True, 12+1j}
print(s)
print(type(s))

print(s[0]) # set does not support indexing
```

```python
# set is growable in nature, based on our requirement we can increase or decrease the size.
s.add(121)
print(s)

# We can remove the element from set
s.remove(10)
print(s)
```

**12) frozenset Data Type:**

֍ It is exactly same as set except that it is immutable.

֍ Hence we cannot use add or remove functions.

֍ .add and .remove methods will not work here

```python
s={10,20,30,40}
fs = frozenset(s)
print(type(fs))
```

```python
fs.add(12)
fs.remove(10)
```

**13) dict Data Type:**

֍ If we want to represent a group of values as key-value pairs then we should go for dict data type.

֍ Eg: d = {101:'Manoj',102:'Digambad',103:'Pankaj'}

֍ Duplicate keys are not allowed but values can be duplicated. If we are trying to insert an entry with duplicate key then old value will be replaced with new value.

֍ dict is mutable and the order won’t be preserved

```python
d = {101:'Manoj',102:'Digambar',103:'Amol'}
d[103] = "343"
d[104] = 'Rohit'
print(d)

# We can create empty dictionary as follows
dict = {}

#We can add key-value pairs as follows
dict[1] = "Apple"
dict['asd'] = "Peru"
dict[123] = "32rrg"
print(dict)
```
