Questions

Basic Data Exploration (Questions 1-3)

Q1. Data Loading and Overview

```python
import pandas as pd

#1. Load the CSV file into a pandas DataFrame
students = pd.read_csv("student_data.csv")

#2. Display the first 10 rows
print("First 10 rows :\n", students.head(10))
print("__________________________________________________________________\n")

#3. Show the shape of the dataset (rows, columns)
print("shape of the dataset:", students.shape)
print("__________________________________________________________________\n")

#4. Display the column names and their data types
print("column names and their data types :\n", students.dtypes)
```

Q2. Summary Statistics


```python
# 1. Calculate and display summary statistics for all numerical columns
print("summary statistics for all numerical columns :\n", students.describe())
print("__________________________________________________________________\n")

# 2. Find the mean, median, and standard deviation of CGPA
cgpa_mean = students['CGPA'].mean()
cgpa_median = students['CGPA'].median()
cgpa_std = students['CGPA'].std()

print(f"Mean CGPA: {cgpa_mean:.2f}")
print(f"Median CGPA: {cgpa_median:.2f}")
print(f"Standard Deviation of CGPA: {cgpa_std:.2f}")
print("__________________________________________________________________\n")

# 3. What is the range (min to max) of AttendancePercent?

min_attendance = students['AttendancePercent'].min()
max_attendance = students['AttendancePercent'].max()

print(f"Minimum attendance percent: {min_attendance}")
print(f"Maximum attendance percent: {max_attendance}")
```

Q3. Missing Data Check


```python
# 1. Check if there are any missing values in the dataset
print("All Missing values in dataset :\n", students.isnull().any())
print("__________________________________________________________________\n")
print("Are there any missing values in the DataFrame?", students.isnull().any().any())

# 2. Count the total number of missing values per column
print("__________________________________________________________________\n")
print("Total numebr of missing values per column :\n",students.isnull().sum())

# 3. Calculate the percentage of missing data for each column
print("__________________________________________________________________\n")
print("percentage of missing data for each column :\n",students.isnull().sum() * 100 / len(students))
```

Filtering and Selection (Questions 4-6)

Q4. Conditional Filtering


```python
# 1. Find all students with CGPA greater than 3.5
print("Students with CGPA > 3.5:\n")
print(students[students['CGPA'] > 3.5])
print("__________________________________________________________________\n")

# 2. How many students have AttendancePercent below 75%?
count_below_75 = (students['AttendancePercent'] < 75).sum()
print("Number of students have AttendancePercent below 75% :", count_below_75)
print("__________________________________________________________________\n")

# 3. Display students from the CS department with FinalScore above 80
filtered_students = students[(students['Department'] == 'CS') & (students['FinalScore'] > 80)]

print("Filtered Students (CS Department, Score > 80):")
print(filtered_students)
```

Q5. Data Selection


```python
# 1. Select only the columns: Name, Department, CGPA, FinalScore
selected_columns = students[['Name', 'Department', 'CGPA', 'FinalScore']]
print("Selected Columns:")
print(selected_columns)
print("__________________________________________________________________\n")

# 2. Extract rows where students are in semester 5 or above
semester_5_or_above = students[students['Semester'] >= 5]
print("Students in semester 5 or above:")
print(semester_5_or_above)
print("__________________________________________________________________\n")

# 3. Get the details of students aged 22 or older from the Math department
math_department_22_or_older = students[(students['Age'] >= 22) & (students['Department'] == 'Math')]
print("Students aged 22 or older from the Math department:")
print(math_department_22_or_older)
```

Q6. Top Performers


```python
# 1. Find the top 10 students based on CGPA
print("Top 10 students based on CGPA:")
print(students.nlargest(10, 'CGPA'))
print("__________________________________________________________________\n")

# 2. Display the bottom 5 students based on FinalScore
print("Bottom 5 students based on FinalScore:")
print(students.nsmallest(5, 'FinalScore'))
print("__________________________________________________________________\n")

# 3. Show students who scored above 90 in both MidtermScore and FinalScore
print("Students who scored above 90 in both MidtermScore and FinalScore:")
print(students[(students['MidtermScore'] > 90) & (students['FinalScore'] > 90)])
```

Aggregation and Grouping (Questions 7-9)

Q7. Department-wise Analysis


```python
# 1. Group data by Department and calculate the average CGPA for each
department_avg_cgpa = students.groupby('Department')['CGPA'].mean()
print("Department-wise Average CGPA:")
print(department_avg_cgpa)
print("__________________________________________________________________\n")

# 2. Find which department has the highest average FinalScore
print("Department with the highest average FinalScore:", students.groupby('Department')['FinalScore'].mean().idxmax())
print("__________________________________________________________________\n")

# 3. Count the number of students in each department
department_student_count = students['Department'].value_counts()
print("Number of students in each department:")
print(department_student_count)
```

Q8. Gender-based Statistics


```python
# 1. Calculate the average AssignmentScore for each Gender
print("Average AssignmentScore for each Gender:")
print(students.groupby('Gender')['AssignmentScore'].mean())
print("__________________________________________________________________\n")

# 2. Find the department with the most Female students
print("Department with the most Female students:", students[students['Gender'] == 'Female'].groupby('Department').size().idxmax())
print("__________________________________________________________________\n")

# 3. Compare the mean AttendancePercent across different Genders
print("Mean AttendancePercent across different Genders:")
print(students.groupby('Gender')['AttendancePercent'].mean())
```

Q9. Advanced Grouping


```python
# 1. Group by Department and Semester, then calculate average ProjectScore
print("Average ProjectScore by grouping Department and Semester:")
print(students.groupby(['Department', 'Semester'])['ProjectScore'].mean())
print("__________________________________________________________________\n")

# 2. Find the semester with the highest average CGPA
print("Semester with the highest average CGPA :", students.groupby('Semester')['CGPA'].mean().idxmax())
print("Semester with the highest average CGPA value :", students.groupby('Semester')['CGPA'].mean().max())
print("__________________________________________________________________\n")

# 3. Create a pivot table showing average FinalScore by Department and Gender
pivot_table_avg_score = pd.pivot_table(students,
                                       values='FinalScore',
                                       index='Department',
                                       columns='Gender',
                                       aggfunc='mean')

print("Pivot Table average FinalScore by Department and Gender:")
print(pivot_table_avg_score)
```

NumPy Operations (Questions 10-11)

Q10. NumPy Array Operations


```python
# Convert the CGPA column to a NumPy array
import numpy as np
cgpa_array = students['CGPA'].to_numpy()
print("CGPA Array:")
print(cgpa_array)
print("__________________________________________________________________\n")

# Calculate the 25th, 50th, and 75th percentiles of all numerical scores
percentiles = students.select_dtypes(include='number').quantile([0.25, 0.50, 0.75])
print("25th, 50th, and 75th percentiles of all numerical scores:")
print(percentiles)
print("__________________________________________________________________\n")

# Use NumPy to calculate the correlation between MidtermScore and FinalScore
correlation_midterm_final = np.corrcoef(students['MidtermScore'], students['FinalScore'])[0, 1]
print("Correlation between MidtermScore and FinalScore:", correlation_midterm_final)
print("__________________________________________________________________\n")
```

Q11. Statistical Analysis with NumPy


```python
# Calculate the total score for each student (sum of AssignmentScore, MidtermScore, FinalScore, ProjectScore)
score_columns = ['AssignmentScore', 'MidtermScore', 'FinalScore', 'ProjectScore']
students['TotalScore'] = students[score_columns].sum(axis=1)

print("Total Score for each student:")
print(students[['Name', 'TotalScore']])
print("__________________________________________________________________\n")

# Use NumPy to find students whose total score is above the 80th percentile
total_scores = students['TotalScore'].values
percentile_80 = np.percentile(total_scores, 80)
above_80th = total_scores > percentile_80
students_above_80th = students[above_80th]

print("Students with Total Score above 80th percentile:")
print(students_above_80th[['Name', 'TotalScore']])
print("__________________________________________________________________\n")


# Calculate the standard deviation of AttendancePercent using NumPy
std_AttendancePercent = np.std(students['AttendancePercent'])
print("Standard Deviation of AttendancePercent:", std_AttendancePercent)
```

Data Manipulation (Question 12)

Q12. Feature Engineering


```python
# 1. Create a new column 'TotalScore' by adding all score columns
total_score_columns = ['AssignmentScore', 'MidtermScore', 'FinalScore', 'ProjectScore']
students['TotalScore'] = students[total_score_columns].sum(axis=1)
print("TotalScore column:")
print(students[['Name', 'TotalScore']])
print("__________________________________________________________________\n")

# 2. Create a 'Grade' column based on CGPA: A (3.5-4.0), B (3.0-3.5), C (2.5-3.0), D (below 2.5)
students['Grade'] = np.where(students['CGPA'] >= 3.5, 'A',
                 np.where(students['CGPA'] >= 3.0, 'B',
                 np.where(students['CGPA'] >= 2.5, 'C', 'D')))

print("Grade Column based on CGPA:")
print(students[['Name', 'CGPA', 'Grade']])
print("__________________________________________________________________\n")

# 3. Create a 'PassStatus' column: 'Pass' if FinalScore >= 50, else 'Fail'
students['PassStatus'] = np.where(students['FinalScore'] >= 50, 'Pass', 'Fail')
print("'PassStatus' column:")
print(students[['Name', 'FinalScore', 'PassStatus']])
print("__________________________________________________________________\n")

# 4. Calculate the percentage of students who passed
pass_percentage = (students['PassStatus'] == 'Pass').mean() * 100
print("Percentage of Students who Passed:", pass_percentage)
print("__________________________________________________________________\n")


# 5. Export the modified DataFrame to a new CSV file called 'student_data_processed.csv'
students.to_csv('student_data_processed.csv', index=False)
df_from_csv = pd.read_csv('student_data_processed.csv')

print("New CSV file:")
print(df_from_csv)
```
